```python
from scipy import stats
import numpy as np
```

```python
stats.invgamma.fit(stats.norm.rvs(loc=1e5, scale=1, size=1000), floc=0, fscale=1)
```

```python
invgamma = stats.invgamma.freeze(a=3, scale=5e5)
points = [invgamma.pdf(i) for i in np.linspace(1e3, 1e6, 100)]
print(invgamma.mean())
c, a, m = toyplot.plot(np.linspace(1e3, 1e6, 100), points, );#yscale="log");
a.vlines(1e5);
```

```python
invgamma = stats.invgamma.freeze(a=3, scale=5e-9)
points = [invgamma.pdf(i) for i in np.linspace(2e-10, 2e-8, 100)]
print(invgamma.mean())
c, a, m = toyplot.plot(np.linspace(2e-10, 2e-8, 100), points, );#yscale="log");
a.vlines(2e-9);
```
